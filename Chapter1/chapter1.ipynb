# **IPython: além do Python normal**

IPython é um ambiente interativo de computação para a linguagem de programação Python. Ele fornece um prompt interativo, onde os usuários podem digitar e executar comandos Python em tempo real, bem como uma série de recursos adicionais que tornam a programação em Python mais eficiente e produtiva.

O IPython oferece recursos como autocompletar código, histórico de comandos, acesso fácil à documentação, depuração interativa, visualização de gráficos e muito mais. Ele também suporta a execução de código em outras linguagens, como Julia, R e Bash, através de "células mágicas".

Além disso, o IPython é usado frequentemente em áreas como ciência de dados, análise de dados, aprendizado de máquina e pesquisa científica, devido à sua capacidade de manipulação interativa e exploração de dados.

Para inciar o ipython, basta digitar o comando `ipython` na linha de comando;

# **Atalhos de teclado no IPython Shell**


* `Ctrl-a`	Move o cursor para o início da linha
* `Ctrl-e`	Mover o cursor para o final da linha
* `Ctrl-bou a tecla de seta para a esquerda`	Mover o cursor um caractere para trás
* `Ctrl-f ou a tecla de seta para a direita`	Mover o cursor um caractere para frente

**Atalhos do histórico de comandos**
Talvez os atalhos mais impactantes discutidos aqui sejam aqueles que o IPython fornece para navegar no histórico de comandos. Este histórico de comandos vai além da sua sessão atual do IPython: todo o seu histórico de comandos é armazenado em um banco de dados SQLite em seu diretório de perfil do IPython. # **Comandos Mágicos ipython**

1. `%paste` Na colagem direta, o intérprete fica confuso com os caracteres de prompt 
adicionais. Mas não tema – a função mágica do IPython `%paste` foi projetada para lidar com esse tipo exato de entrada marcada com várias linhas;

2. `cpaste` Um comando com intenção semelhante é %cpaste, que abre um prompt interativo de várias linhas no qual você pode colar um ou mais blocos de código a serem executados em lote;

3. `%run` comando que serve para executar algum script com o código previamente editado. Exemplo: 

arquivo: myscript.py
```
#arquivo myscript.py
def  quadrado ( x ): 
    """quadrar um número""" 
    return  x  **  2

para  N  no  intervalo ( 1 ,  4 ): 
    print ( N ,  "quadrado é" ,  quadrado ( N ))

```
`% run myscript.py`




4. Outro exemplo de uma função mágica útil é `%timeit`, que determinará automaticamente o tempo de execução da instrução Python de linha única que a segue.

# Comandos IPython e Shell

`!` O ponto de exclamação, quando adicionado na frente de qualquer comando, faz com que uma linha será executada não pelo kernel do Python, mas pela linha de comando do sistema.

abaixo está um exemplo onde um usuário explora alguns comandos basicos do shel do linux:

```
osx:~ $ echo  "hello world"              # echo é como a função de impressão do Python
Olá Mundo

osx:~ $ pwd                             # pwd = print diretório de trabalho 
/home/jake                              # este é o "caminho" em que estamos

osx:~ $ ls                              # ls = lista o conteúdo do diretório de trabalho
projetos de cadernos

osx:~ $ cd projects/                    # cd = alterar diretório

osx:projetos $ pwd
/home/jake/projetos

osx:projetos $ ls
datasci_book mpld3 meuprojeto.txt

osx:projects $ mkdir meuprojeto           # mkdir = criar novo diretório

osx:projetos $ cd meuprojeto/

osx:meuprojeto $ mv ../meuprojeto.txt ./   # mv = mover arquivo. Aqui estamos movendo o 
                                        # arquivo myproject.txt de um diretório 
                                        # para cima (../) para o diretório atual (./)
osx:meuprojeto $ls
meuprojeto.txt

```

**Depuração: quando a leitura de tracebacks não é suficiente**

A ferramenta Python padrão para depuração interativa é pdb, o depurador Python. Este depurador permite que o usuário percorra o código linha por linha para ver o que pode estar causando um erro mais difícil. A versão aprimorada do IPython disso é ipdb, o depurador do IPython.

No IPython, talvez a interface mais conveniente para depuração seja o %debug comando mágico. Se você chamá-lo após atingir uma exceção, ele abrirá automaticamente um prompt de depuração interativo no ponto da exceção. O ipdb prompt permite explorar o estado atual da pilha, explorar as variáveis ​​disponíveis e até mesmo executar comandos Python!

image.png


Por fim, se você tiver um script que gostaria de executar desde o início no modo interativo, poderá executá-lo com o comando %run -d e usar o nextcomando para percorrer as linhas de código interativamente.
