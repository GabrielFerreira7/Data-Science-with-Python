# Noções básicas de matrizes NumPy

Abordaremos algumas categorias de manipulações básicas de array aqui:

* Atributos de arrays : determinando o tamanho, forma, consumo de memória e tipos de dados de arrays;
* Indexação de arrays : obtendo e definindo o valor de elementos individuais do array;
* Fatiamento de arrays : obtendo e configurando subarrays menores dentro de um array maior;
* Remodelagem de matrizes : alterando a forma de uma determinada matriz;
* Junção e divisão de arrays : Combinando vários arrays em um e dividindo um array em muitos.

````python
'''
Para exemplificar, iremos definir 3 arrays aleatório no numpay, um unidimensional, bidimensional e tridimensional
'''
import numpy as np
np.random.seed(0)  # seed for reproducibility

x1 = np.random.randint(10, size=6)  # One-dimensional array
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array

'''
Temos alguns atributos referente a cada array, como ndim(o número de dimensões), shape(o tamanho de cada dimensão) e size(o tamanho total do array)
que pode ser obtido através dos seguintes comandos
'''

print("x3 ndim: ", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size)
````

````python
'Em uma matriz unidimensional o array pode ser acessado através dos indice indicado entre colchetes'

print(x1[0])
print(x1[4])


'Em uma matriz multidimensional, os itens podem ser acessados ​​usando uma tupla de índices separados por vírgulas:'

print("acessando matriz multidimensional")
print(x2 [ 0 ,  0 ])
print(x2 [ 2 ,  - 1 ])
````

**Fatiamento de Array: Acessando Subarrays **

Assim como podemos usar colchetes para acessar elementos individuais do array, também podemos usá-los para acessar subarrays com a notação de fatia , marcada pelo :caractere de dois pontos ( ). A sintaxe de divisão do NumPy segue a da lista padrão do Python; para acessar uma fatia de um array x, use isto:


```
x[start:stop:step]
```



````python
x = np.arange(10)
print(x)

print(x[:5])  # primeiros 5 elementos

print(x[5:])  # elementos após o indice 5

print(x[4:7])  # sub array do meio 

print(x[::2])  # todos os elementos indo de 2 em 2

print(x[1::2])  # elementos indo de 2 em 2 começando o elemento indice 1
````

**Subarrays multidimensionais** 

As fatias multidimensionais funcionam da mesma maneira, com várias fatias separadas por vírgulas. Por exemplo:



````python
print(x2)
print(x2 [:2 , :3])  # duas linhas, três colunas)

print(x2 [: 3 ,  :: 2 ])   # todas as linhas, colunas alternadas)
````

Todas as rotinas anteriores funcionaram em arrays únicos. Também é possível combinar várias matrizes em uma e, inversamente, dividir uma única matriz em várias matrizes. Vamos dar uma olhada nessas operações aqui.

A concatenação, ou junção de dois arrays no NumPy, é realizada principalmente usando as rotinas `np.concatenate`, `np.vstack` `np.hstack`. np.concatenaterecebe uma tupla ou lista de arrays como seu primeiro argumento, como podemos ver aqui:

````python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
print(np.concatenate([x, y]))

z = [99, 99, 99]
print(np.concatenate([x, y, z]))
````

````python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])

# concatenate along the first axis
print(np.concatenate([grid, grid]))

# concatenate along the second axis (zero-indexed)
print(np.concatenate([grid, grid], axis=1))
````

Para trabalhar com arrays de dimensões mistas, pode ser mais fácil usar as funções `np.vstack`(pilha vertical) e (pilha horizontal):`np.hstack`

````python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],
                 [6, 5, 4]])

# vertically stack the arrays
print("vertical")
print(np.vstack([x, grid]))

print("Horizontal")

# horizontally stack the arrays
y = np.array([[99],
              [99]])
print(np.hstack([grid, y]))
````

**Divisão de arrays** 
O oposto da concatenação é a divisão, que é implementada pelas funções `np.split`, `np.hsplit` `np.vsplit`. Para cada um deles, podemos passar uma lista de índices dando os pontos de divisão:



````python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
````

````python
grid = np.arange(16).reshape((4, 4))
grid
````

````python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
````

````python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
````

# Computação em matrizes NumPy: funções universais

A computação em matrizes NumPy pode ser muito rápida ou muito lenta. A chave para torná-lo rápido é usar operações vetorizadas , geralmente implementadas por meio das funções universais do NumPy (ufuncs). Esta seção motiva a necessidade dos ufuncs do NumPy, que podem ser usados ​​para fazer cálculos repetidos em elementos de matriz muito mais eficientes. Em seguida, apresenta muitos dos ufuncs aritméticos mais comuns e úteis disponíveis no pacote NumPy.





Operações vetorizadas em NumPy são implementadas via ufuncs , cujo principal objetivo é executar rapidamente operações repetidas em valores em arrays NumPy. Ufuncs são extremamente flexíveis.

abaixo um exemplo de operações sem vetorização e uma com vetorização

````python
'''
Imagine que vc precise todos elementos de um array por 100
Um trecho de código comum que faria esse processo seria o seguinte

'''
x = np.arange(9, dtype=numpy.float32)
print(x)

for i in range(0, x.size):
  x[i] = x[i]/100

print(x)

````

Para muitos tipos de operações, o NumPy fornece uma interface conveniente para esse tipo de rotina compilada estaticamente digitada. Isso é conhecido como uma operação vetorizada . Isso pode ser feito simplesmente executando uma operação no array, que será aplicada a cada elemento. Essa abordagem vetorizada foi projetada para enviar o loop para a camada compilada subjacente ao NumPy, levando a uma execução muito mais rápida.
O mesmo procedimento de divisão, pode ser feito de uma maneira vetorizada.

````python
x2 = np.arange(9, dtype=numpy.float32)
print(x2)
x2 = x2/100
print(x2)
````

# Computação em Arrays: Broadcasting

Broadcasting é simplesmente um conjunto de regras para aplicar ufuncs binários (por exemplo, adição, subtração, multiplicação, etc.) em arrays de tamanhos diferentes.

Para arrays do mesmo tamanho, as operações binárias são executadas elemento por elemento.
Com o broadcasting, é possível fazer operaões em arrays com tamanho distintos, observe o seguinte exemplo.

````python
a = np.array([0, 1, 2])
print("ARRAY a: ")
print(a)

#operação de Broadcasting
a = a + 6
print("ARRAY a depois do Broadcasting: ")
print(a)
````

odemos pensar nisso como uma operação que estende ou duplica o valor 5na matriz `[5, 5, 5]` e adiciona os resultados. A vantagem da transmissão do NumPy é que essa duplicação de valores não ocorre realmente, mas é um modelo mental útil quando pensamos em Broadcasting.

O mesmo é válido para estruturas multidimensionais, como pode ser observado no exemplo abaixo

````python
a = np.array([1, 4, 5])
b = np.zeros((3,3)) + 4
print("Array a")
print(a)

print("Array b")
print(b)

print("Array b com Broadcasting")
print(a+b)
````

O Broadcasting no NumPy segue um conjunto estrito de regras para determinar a interação entre os dois arrays:

* Regra 1: Se as duas matrizes diferem em seu número de dimensões, a forma daquela com menos dimensões é preenchida com as do lado esquerdo (à esquerda).
* Regra 2: Se a forma das duas matrizes não corresponder em nenhuma dimensão, a matriz com forma igual a 1 nessa dimensão é alongada para corresponder à outra forma.
*Regra 3: Se em alguma dimensão os tamanhos discordarem e nenhum for igual a 1, um erro é levantado.

````python
#Exemplo de transmissão 1
M = np.ones((2, 3))
a = np.arange(3)

print(M + a)
````

Vamos considerar uma operação nesses dois arrays. A forma das matrizes são

```
M.shape = (2, 3)
a.shape = (3,)
```


Vemos pela regra 1 que o array atem menos dimensões, então o preenchemos à esquerda com um:
```
M.shape -> (2, 3)
a.shape -> (1, 3)
```
Pela regra 2, agora vemos que a primeira dimensão discorda, então esticamos essa dimensão para corresponder:
```
M.shape -> (2, 3)
a.shape -> (2, 3)
```
As formas coincidem e vemos que a forma final será (2, 3):

Agora vamos dar uma olhada em um exemplo em que os dois arrays não são compatíveis:

`M  =  np . ones (( 3 ,  2 )) `
`a  =  np.arange ( 3 )`
Esta é apenas uma situação ligeiramente diferente do primeiro exemplo: a matriz Mé transposta. Como isso afeta o cálculo? A forma das matrizes são

`M.shape = (3, 2)`
`a.shape = (3,)`
Novamente, a regra 1 nos diz que devemos preencher a forma de acom uns:

`M.shape -> (3, 2)`
`a.shape -> (1, 3)`
Pela regra 2, a primeira dimensão de aé estendida para coincidir com a de M:

`M.shape -> (3, 2)`
`a.shape -> (3, 3)`
Agora atingimos a regra 3 – as formas finais não combinam, então essas duas matrizes são incompatíveis, como podemos observar ao tentar esta operação:

**À parte: Usando as palavras-chave e/ou versus os operadores &/|**

A diferença é esta: ande oravaliam a veracidade ou falsidade de todo o objeto , enquanto `&` e `|`referem-se a bits dentro de cada objeto .

Quando você usa andou or, é equivalente a pedir ao Python para tratar o objeto como uma única entidade booleana. Em Python, todos os inteiros diferentes de zero serão avaliados como True


Quando você usa `&` e `|` em números inteiros, a expressão opera nos bits do elemento, aplicando o e ou o ou aos bits individuais que compõem o número.


Portanto, lembre-se disso: ande orexecute uma única avaliação booleana em um objeto inteiro, enquanto &e |execute várias avaliações booleanas no conteúdo (os bits ou bytes individuais) de um objeto. Para arrays Boolean NumPy, o último é quase sempre a operação desejada.

# Ordenação de Arrays



A ordenação de vetores é uma tarefa comum na ciência de dados e pode ter várias utilidades, dependendo do contexto do problema em questão. Podemos citar as seguintes utilidades:

* Busca: A ordenação é uma etapa importante na busca por um elemento específico em um vetor. Se o vetor não estiver ordenado, pode ser necessário percorrer todo o vetor para encontrar o elemento desejado. No entanto, se o vetor estiver ordenado, é possível utilizar algoritmos de busca mais eficientes, como a busca binária, que pode encontrar o elemento desejado em tempo logarítmico em relação ao tamanho do vetor.

* Análise de dados: A ordenação pode ser útil para analisar padrões e tendências em conjuntos de dados. Por exemplo, se um conjunto de dados estiver ordenado por data ou tempo, é possível visualizar a evolução de uma variável ao longo do tempo.

* Redução de complexidade: Em muitos casos, a ordenação pode reduzir a complexidade de um algoritmo. Por exemplo, se um algoritmo precisa encontrar o maior elemento em um vetor, pode ser necessário percorrer todo o vetor para encontrar o elemento desejado. No entanto, se o vetor estiver ordenado em ordem decrescente, o maior elemento estará na primeira posição do vetor.

* Algoritmos de classificação: A ordenação é uma etapa essencial em muitos algoritmos de classificação, como o quicksort e o mergesort. Esses algoritmos usam a ordenação para dividir o vetor em subconjuntos menores e, em seguida, classificar cada subconjunto separadamente.

É possível utilizar a ordenação no numpy, chamando o método `np.sort`. Por padrão, o numpy oferece a implementação do QuickSort, que possui uma complexidade O(NlogN) e é o mais eficiente para a maioria das aplicações 

````python
x = np.array([2, 1, 4, 3, 5])
np.sort(x)
````

**Ordenação ao longo de linhas e colunas**

Um recurso útil dos algoritmos de classificação do NumPy é a capacidade de classificar linhas ou colunas específicas de uma matriz multidimensional usando o `axis` argumento. Por exemplo:

````python
rand = np.random.RandomState(42)
X = rand.randint(0, 10, (4, 6))
print(X)
print("ordenando no eixo 0, mantendo os elementos da coluna ordenada")
print(np.sort(X, axis=0))

print("ordenando no eixo 1, mantendo os elementos da linha ordenada")
print(np.sort(X, axis=1))
````

**Classificações Parciais: Particionamento **
Às vezes, não estamos interessados ​​em classificar todo o array, mas simplesmente em encontrar os k menores valores no array. NumPy fornece isso na np.partitionfunção. np.partitionrecebe um array e um número K ; o resultado é uma nova matriz com os menores valores de K à esquerda da partição e os valores restantes à direita, em ordem arbitrária:

Observe que os três primeiros valores na matriz resultante são os três menores na matriz e as posições restantes da matriz contêm os valores restantes. Dentro das duas partições, os elementos têm ordem arbitrária.

````python
x = np.array([7, 2, 3, 1, 6, 5, 4])
np.partition(x, 3)
````

Um bom exemplo seria o K-vizinhos mais proximos que pode ser implementado com o `np.partition`

# Dados estruturados: matrizes estruturadas do NumPy

Imagine que temos várias categorias de dados sobre várias pessoas (digamos, nome, idade e peso) e gostaríamos de armazenar esses valores para uso em um programa Python. Seria possível armazená-los em três matrizes separadas.

Mas isso é um pouco desajeitado. Não há nada aqui que nos diga que as três matrizes estão relacionadas; seria mais natural se pudéssemos usar uma única estrutura para armazenar todos esses dados. O NumPy pode lidar com isso por meio de matrizes estruturadas, que são matrizes com tipos de dados compostos.

Lembre-se de que anteriormente criamos um array simples usando uma expressão como esta:
`x = np.zeros(4, dtype=int)`

Da mesma forma, podemos criar uma matriz estruturada usando uma especificação de tipo de dados compostos:

````python
# Use a compound data type for structured arrays
data = np.zeros(4, dtype={'names':('name', 'age', 'weight'),
                          'formats':('U10', 'i4', 'f8')})
print(data.dtype)
````

Aqui `'U10'`se traduz em "string Unicode de comprimento máximo 10", `'i4'`se traduz em "número inteiro de 4 bytes (ou seja, 32 bits)" e `'f8'`se traduz em "float de 8 bytes (ou seja, 64 bits)

````python
data['name'] = ['Alice', 'Bob', 'Cathy', 'Doug']
data['age'] = [25, 45, 37, 19]
data['weight'] = [55.0, 85.5, 68.0, 61.5]
print(data)
````

Usando o mascaramento booleano, isso ainda permite que você faça algumas operações mais sofisticadas, como filtragem por idade:

````python
# Get names where age is under 30
data[data['age'] < 30]['name']
````
