## Categorias de Machine Learning
No nível mais fundamental, o aprendizado de máquina pode ser categorizado em dois tipos principais: aprendizado supervisionado e aprendizado não supervisionado.

1) O aprendizado supervisionado envolve, de alguma forma, modelar o relacionamento entre os recursos medidos dos dados e algum rótulo associado aos dados; uma vez determinado esse modelo, ele pode ser usado para aplicar rótulos a dados novos e desconhecidos. Isso é ainda subdividido em tarefas de classificação e tarefas de regressão : na classificação, os rótulos são categorias discretas, enquanto na regressão, os rótulos são quantidades contínuas.

2) O aprendizado não supervisionado envolve a modelagem dos recursos de um conjunto de dados sem referência a nenhum rótulo e é frequentemente descrito como "deixar o conjunto de dados falar por si". Esses modelos incluem tarefas como agrupamento e redução de dimensionalidade. Os algoritmos de agrupamento identificam grupos distintos de dados, enquanto os algoritmos de redução de dimensionalidade buscam representações mais sucintas dos dados.

O aprendizado de máquina trata da criação de modelos a partir de dados: por esse motivo, começaremos discutindo como os dados podem ser representados para serem compreendidos pelo computador. A melhor maneira de pensar sobre os dados no Scikit-Learn é em termos de tabelas de dados.

##Dados como tabela
Uma tabela básica é uma grade bidimensional de dados, na qual as linhas representam elementos individuais do conjunto de dados e as colunas representam quantidades relacionadas a cada um desses elementos. Por exemplo, considere o conjunto de dados Iris , analisado por Ronald Fisher em 1936. Podemos baixar este conjunto de dados na forma de um Pandas DataFrameusando a biblioteca seaborn:

````python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
````

Aqui, cada linha dos dados refere-se a uma única flor observada e o número de linhas é o número total de flores no conjunto de dados. Em geral, nos referiremos às linhas da matriz como amostras e ao número de linhas como `n_samples`.

Da mesma forma, cada coluna de dados refere-se a uma informação quantitativa específica que descreve cada amostra. Em geral, nos referiremos às colunas da matriz como recursos , e ao número de colunas como `n_features`.

### Matriz de características
Esse layout de tabela deixa claro que as informações podem ser pensadas como uma matriz ou matriz numérica bidimensional, que chamaremos de matriz de características . Por convenção, essa matriz de características é frequentemente armazenada em uma variável chamada X. Supõe-se que a matriz de recursos seja bidimensional, com forma `[n_samples, n_features]`

###Matriz de destino
Além da matriz de recursos X, geralmente também trabalhamos com um rótulo ou array de destino , que por convenção costumamos chamar de y. A matriz de destino geralmente é unidimensional, com length n_samples, e geralmente está contida em uma matriz NumPy ou Pandas Series. A matriz de destino pode ter valores numéricos contínuos ou classes/rótulos discretos. Embora alguns estimadores do Scikit-Learn lidem com vários valores de destino na forma de uma matriz de destino bidimensional `[n_samples, n_targets]`, trabalharemos principalmente com o caso comum de uma matriz de destino unidimensional.

Para uso no `Scikit-Learn`, extrairemos a matriz de recursos e a matriz de destino do `DataFrame`, o que podemos fazer usando algumas das DataFrameoperações do Pandas

````python
X_iris = iris.drop('species', axis=1)
X_iris.shape
````

````python
y_iris = iris['species']
y_iris.shape
````

##Exemplo de regressão linear


````python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
````

Com esses dados no lugar, podemos usar a receita descrita anteriormente. Vamos percorrer o processo:

1. Escolha uma classe de modelo
No Scikit-Learn, cada classe de modelo é representada por uma classe Python. Assim, por exemplo, se quisermos calcular um modelo de regressão linear simples, podemos importar a classe de regressão linear:

````python
from sklearn.linear_model import LinearRegression
````

2) Escolha os hiperparâmetros do modelo
Um ponto importante é que uma classe de modelo não é o mesmo que uma instância de um modelo .

Depois de decidirmos nossa classe de modelo, ainda há algumas opções em aberto para nós. Dependendo da classe de modelo com a qual estamos trabalhando, talvez seja necessário responder a uma ou mais perguntas como as seguintes:

* Gostaríamos de ajustar o deslocamento (isto é, interceptação y )?
* Gostaríamos que o modelo fosse normalizado?
* Gostaríamos de pré-processar nossos recursos para adicionar flexibilidade ao modelo?
* Que grau de regularização gostaríamos de usar em nosso modelo?
* Quantos componentes de modelo gostaríamos de usar?

Estes são exemplos de escolhas importantes que devem ser feitas uma vez que a classe de modelo é selecionada. Essas opções geralmente são representadas como hiperparâmetros ou parâmetros que devem ser definidos antes que o modelo seja ajustado aos dados. No Scikit-Learn, os hiperparâmetros são escolhidos passando valores na instanciação do modelo.



Para nosso exemplo de regressão linear, podemos instanciar a LinearRegressionclasse e especificar que gostaríamos de ajustar a interceptação usando o `fit_intercept` hiperparâmetro:

````python
model = LinearRegression(fit_intercept=True)
model
````

3) Organize os dados em uma matriz de recursos e vetor de destino


Anteriormente, detalhamos a representação de dados do Scikit-Learn, que requer uma matriz de recursos bidimensionais e uma matriz de destino unidimensional. Aqui, nossa variável de destino yjá está na forma correta (uma `n_samplesmatriz` de comprimento), mas precisamos massagear os dados xpara torná-la uma matriz de tamanho `[n_samples, n_features]`. Nesse caso, isso equivale a uma simples reformulação da matriz unidimensional:

````python
X = x[:, np.newaxis]
X.shape
````

4) Ajuste o modelo aos seus dados ¶
Agora é hora de aplicar nosso modelo aos dados. Isso pode ser feito com o fit()método do modelo:

````python
model.fit(X, y)
````

Esse `fit()` comando faz com que vários cálculos internos dependentes do modelo ocorram e os resultados desses cálculos são armazenados em atributos específicos do modelo que o usuário pode explorar. No Scikit-Learn, por convenção, todos os parâmetros do modelo que foram aprendidos durante o `fit()`processo têm sublinhados à direita; por exemplo, neste modelo linear, temos o seguinte

````python
model.coef_
````

````python
model.intercept_
````

Esses dois parâmetros representam a inclinação e a interceptação do ajuste linear simples aos dados. Comparando com a definição dos dados, vemos que eles estão muito próximos da inclinação de entrada de 2 e interceptação de -1.

Uma questão que frequentemente surge diz respeito à incerteza em tais parâmetros internos do modelo. Em geral, o Scikit-Learn não fornece ferramentas para tirar conclusões dos próprios parâmetros internos do modelo: interpretar os parâmetros do modelo é muito mais uma questão de modelagem estatística do que uma questão de aprendizado de máquina . O aprendizado de máquina se concentra no que o modelo prevê .

5) Prever rótulos para dados desconhecidos


Depois que o modelo é treinado, a principal tarefa do aprendizado de máquina supervisionado é avaliá-lo com base no que ele diz sobre novos dados que não faziam parte do conjunto de treinamento. No Scikit-Learn, isso pode ser feito usando o `predict()` método. Para fins deste exemplo, nossos "novos dados" serão uma grade de valores x e perguntaremos quais valores y o modelo prevê:

````python
xfit = np.linspace(-1, 11)
````

Como antes, precisamos coagir esses valores x em uma `[n_samples, n_features]`matriz de recursos, após a qual podemos alimentá-la ao modelo:

````python
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
````

````python
plt.scatter(x, y)
plt.plot(xfit, yfit);
````

##Exemplo de aprendizado supervisionado: classificação de íris

Vamos dar uma olhada em outro exemplo desse processo, usando o conjunto de dados Iris que discutimos anteriormente. Nossa pergunta será esta: dado um modelo treinado em uma parte dos dados do Iris, quão bem podemos prever os rótulos restantes?




Gostaríamos de avaliar o modelo em dados que ele não viu antes e, portanto, dividiremos os dados em um conjunto de treinamento e um conjunto de teste . Isso pode ser feito manualmente, mas é mais conveniente usar a `train_test_split` função de utilidade:

````python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
````

Com os dados organizados, podemos seguir nossa receita para prever os rótulos:

````python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
````

Por fim, podemos usar o `accuracy_score` utilitário para ver a fração de rótulos previstos que correspondem ao seu valor verdadeiro:

````python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
````

#Engenharia de recursos
 No mundo real, os dados raramente vêm na forma `[n_samples, n_features]`. Com isso em mente, uma das etapas mais importantes no uso do aprendizado de máquina na prática é a engenharia de recursos : ou seja, pegar qualquer informação que você tenha sobre seu problema e transformá-la em números que você pode usar para construir sua matriz de recursos.

##Características categóricas
Um tipo comum de dados não numéricos são os dados categóricos . Por exemplo, imagine que você está explorando alguns dados sobre preços de moradias e, junto com características numéricas como "preço" e "quartos", você também tem informações de "vizinhança". Por exemplo, seus dados podem ter a seguinte aparência:

````python
data = [
    {'price': 850000, 'rooms': 4, 'neighborhood': 'Queen Anne'},
    {'price': 700000, 'rooms': 3, 'neighborhood': 'Fremont'},
    {'price': 650000, 'rooms': 3, 'neighborhood': 'Wallingford'},
    {'price': 600000, 'rooms': 2, 'neighborhood': 'Fremont'}
]
````

Nesse caso, uma técnica comprovada é usar codificação one-hot , que efetivamente cria colunas extras indicando a presença ou ausência de uma categoria com um valor de 1 ou 0, respectivamente. Quando seus dados vierem como uma lista de dicionários, o Scikit-Learn `DictVectorizer` fará isso por você:

````python
from sklearn.feature_extraction import DictVectorizer
vec = DictVectorizer(sparse=False, dtype=int)
vec.fit_transform(data)
````

Observe que a coluna 'vizinhança' foi expandida em três colunas separadas, representando os três rótulos de vizinhança, e que cada linha tem um 1 na coluna associada à sua vizinhança. Com esses recursos categóricos assim codificados, você pode prosseguir normalmente com o ajuste de um modelo Scikit-Learn.

Para ver o significado de cada coluna, você pode inspecionar os nomes dos recursos:

````python
vec.get_feature_names_out()
````

##Imputação de dados ausentes
Outra necessidade comum na engenharia de recursos é o tratamento de dados ausentes. Discutimos a manipulação de dados ausentes em DataFrames em Manipulando dados ausentes e vimos que frequentemente o `NaN` valor é usado para marcar valores ausentes. Por exemplo, podemos ter um conjunto de dados semelhante a este:

````python
from numpy import nan
X = np.array([[ nan, 0,   3  ],
              [ 3,   7,   9  ],
              [ 3,   5,   2  ],
              [ 4,   nan, 6  ],
              [ 8,   8,   1  ]])
y = np.array([14, 16, -1,  8, -5])
````

Ao aplicar um modelo típico de aprendizado de máquina a esses dados, primeiro precisaremos substituir esses dados ausentes por algum valor de preenchimento apropriado. Isso é conhecido como imputação de valores ausentes, e as estratégias variam de simples (por exemplo, substituição de valores ausentes pela média da coluna) a sofisticadas (por exemplo, uso de preenchimento de matriz ou um modelo robusto para lidar com esses dados).

 Para uma abordagem de imputação de linha de base, usando o valor médio, mediano ou mais frequente, o Scikit-Learn fornece a Imputerclasse:

````python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='mean')
X2 = imp.fit_transform(X)
X2
````

Vemos que nos dados resultantes, os dois valores ausentes foram substituídos pela média dos valores restantes na coluna. Esses dados imputados podem ser alimentados diretamente, por exemplo, em um LinearRegressionestimador:

````python
model = LinearRegression().fit(X2, y)
model.predict(X2)
````

#Em Profundidade: Classificação Naive Bayes

Os modelos Naive Bayes são um grupo de algoritmos de classificação extremamente rápidos e simples que geralmente são adequados para conjuntos de dados de dimensões muito altas. Por serem tão rápidos e terem tão poucos parâmetros ajustáveis, eles acabam sendo muito úteis como uma linha de base rápida e suja para um problema de classificação.

Classificadores Naive Bayes são construídos em métodos de classificação Bayesiana. Estes dependem do teorema de Bayes, que é uma equação que descreve a relação de probabilidades condicionais de quantidades estatísticas. Na classificação bayesiana, estamos interessados em encontrar a probabilidade de um rótulo dados alguns recursos observados, que podemos escrever como P(L | recursos).

Neste classificador, a suposição é que os dados de cada rótulo são extraídos de uma distribuição gaussiana simples . Imagine que você tenha os seguintes dados:

````python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
````

Uma maneira extremamente rápida de criar um modelo simples é assumir que os dados são descritos por uma distribuição gaussiana sem covariância entre as dimensões. Esse modelo pode ser ajustado simplesmente encontrando a média e o desvio padrão dos pontos dentro de cada rótulo, que é tudo o que você precisa para definir essa distribuição. O resultado dessa suposição gaussiana ingênua é mostrado na figura a seguir:

##Resumo

Naive Bayes é um algoritmo de aprendizado de máquina amplamente utilizado para tarefas de classificação. Ele é baseado no teorema de Bayes, que descreve a probabilidade de um evento com base no conhecimento prévio das condições que podem estar relacionadas ao evento.

No Naive Bayes, é feita a suposição "ingênua" de que todas as características (ou atributos) são independentes entre si, dado o rótulo da classe. Apesar dessa suposição simplificadora, o Naive Bayes frequentemente apresenta um bom desempenho na prática e é computacionalmente eficiente. Ele é particularmente eficaz em situações em que a suposição de independência se mantém razoavelmente bem, como em tarefas de classificação de texto.

Aqui está uma visão geral de como o Naive Bayes funciona:

**Fase de Treinamento:**

1. Coletar um conjunto de dados de treinamento rotulados, consistindo de instâncias e suas respectivas classes.
2. Calcular as probabilidades a priori de cada classe com base na frequência com que cada classe aparece no conjunto de treinamento.
3. Calcular as probabilidades condicionais de cada atributo para cada classe, estimando a probabilidade de um determinado valor do atributo dado que a classe seja verdadeira.


**Fase de Teste**:

1. Receber um novo exemplo (instância) não rotulado para classificação.
2. Para cada classe possível, calcular a probabilidade posterior usando o teorema de Bayes e as probabilidades calculadas na fase de treinamento.
3. Atribuir a classe com a maior probabilidade posterior como a classe prevista para o exemplo.

É importante ressaltar que o Naive Bayes é sensível à presença de atributos irrelevantes ou correlacionados, pois a suposição de independência condicional pode ser violada nessas situações. No entanto, mesmo com essa limitação, o Naive Bayes é amplamente utilizado e demonstrou ser eficaz em várias aplicações, como classificação de documentos, detecção de spam, análise de sentimentos e muito mais.

##Support Vector Machines: maximizando a margem

As máquinas de vetores de suporte oferecem uma maneira de melhorar isso. A intuição é esta: em vez de simplesmente traçar uma linha de largura zero entre as classes, podemos traçar em torno de cada linha uma margem de alguma largura, até o ponto mais próximo. Aqui está um exemplo de como isso pode parecer:

````python

xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
````

Nas máquinas de vetores de suporte, a linha que maximiza essa margem é a que escolheremos como modelo ótimo. As máquinas de vetores de suporte são um exemplo desse estimador de margem máxima.

##Montando uma máquina de vetores de suporte
Vamos ver o resultado de um ajuste real a esses dados: usaremos o classificador de vetor de suporte do Scikit-Learn para treinar um modelo SVM nesses dados. Por enquanto, usaremos um kernel linear e definiremos o `C` parâmetro para um número muito grande.

````python
from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
````

Para visualizar melhor o que está acontecendo aqui, vamos criar uma função de conveniência rápida que traçará os limites de decisão do SVM para nós:

````python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)

    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])

    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
````

````python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model);
````

Esta é a linha divisória que maximiza a margem entre os dois conjuntos de pontos. Observe que alguns dos pontos de treinamento apenas tocam a margem. Esses pontos são os elementos centrais desse ajuste e são conhecidos como vetores de suporte e dão nome ao algoritmo. No Scikit-Learn, a identidade desses pontos é armazenada no support_vectors_atributo do classificador:

````python
model.support_vectors_
````

##Além dos limites lineares: Kernel SVM

Nos modelos SVM, podemos usar uma versão da mesma ideia utilizada na regressão de base, em que foi possível representat comportamento não lineares através de regressão linear. Para motivar a necessidade de kernels, vejamos alguns dados que não são linearmente separáveis:

###Ajustando o SVM: suavizando margens

Nossa discussão até agora centrou-se em conjuntos de dados muito limpos, nos quais existe um limite de decisão perfeito. Mas e se seus dados tiverem alguma sobreposição? Por exemplo, você pode ter dados como este:

````python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=1.2)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
````

Para lidar com esse caso, a implementação do SVM possui um fator de correção que "suaviza" a margem: ou seja, permite que alguns dos pontos se infiltrem na margem se isso permitir um ajuste melhor. A dureza da margem é controlada por um parâmetro de ajuste, mais conhecido como `C`
. Para muito grande `C`
, a margem é dura e os pontos não podem estar nela. Para menores `C`
, a margem é mais suave, podendo crescer até englobar alguns pontos.

````python
X, y = make_blobs(n_samples=100, centers=2,
                  random_state=0, cluster_std=0.8)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi, C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel='linear', C=C).fit(X, y)
    axi.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
                model.support_vectors_[:, 1],
                s=300, lw=1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(C), size=14)
````

O valor ótimo do `C`
O parâmetro dependerá do seu conjunto de dados e deve ser ajustado usando validação cruzada ou um procedimento semelhante.

#Em Profundidade: Árvores de Decisão e Florestas Aleatórias

Florestas aleatórias são um exemplo de um conjuntométodo, o que significa que ele se baseia na agregação dos resultados de um conjunto de estimadores mais simples. O resultado um tanto surpreendente com tais métodos de conjunto é que a soma pode ser maior que as partes: ou seja, uma votação majoritária entre vários estimadores pode acabar sendo melhor do que qualquer um dos estimadores individuais fazendo a votação!

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
````

Árvores de decisão são um modelo de aprendizado de máquina supervisionado utilizado para tomar decisões ou prever valores em uma determinada situação. Essas árvores são estruturas hierárquicas que consistem em nós (ou "pontos de decisão") e arestas (ou "ramos") que conectam os nós.

Cada nó interno na árvore representa uma decisão baseada em um atributo ou característica dos dados, enquanto os nós folha representam as classes ou valores previstos. Os nós são divididos com base nas melhores características que separam as classes ou otimizam a previsão dos valores.

O processo de construção de uma árvore de decisão envolve a seleção da melhor característica em cada nó para dividir os dados e a criação de subárvores com base nessa divisão. O algoritmo usado para construir a árvore de decisão pode variar, mas alguns dos métodos populares incluem o algoritmo ID3, o algoritmo C4.5 e o algoritmo CART.

Uma vez construída a árvore de decisão, ela pode ser usada para tomar decisões ou prever valores para novos dados. Isso é feito seguindo o caminho da raiz até um nó folha, onde a classe ou valor previsto é encontrado.

As árvores de decisão têm a vantagem de serem interpretáveis e explicáveis, permitindo que os usuários entendam o raciocínio por trás das decisões tomadas. Elas também podem lidar com diferentes tipos de dados, como numéricos e categóricos, e são eficientes em termos de tempo de treinamento e predição.

No entanto, as árvores de decisão também têm algumas limitações. Elas podem ser sensíveis a pequenas variações nos dados de treinamento, levando a problemas de sobreajuste (overfitting). Além disso, em conjuntos de dados complexos, as árvores de decisão podem se tornar muito grandes e complexas, o que pode dificultar sua interpretação.

Para lidar com essas limitações, técnicas como poda (pruning), limite de profundidade máxima da árvore e uso de florestas aleatórias (random forests) podem ser aplicadas.

Em resumo, as árvores de decisão são um modelo de aprendizado de máquina poderoso e amplamente utilizado, adequado para uma variedade de problemas de classificação e regressão. Elas oferecem interpretabilidade e eficiência, mas exigem cuidados para evitar problemas de sobreajuste e complexidade excessiva.

##Criando uma árvore de decisão
Considere os seguintes dados bidimensionais, que possuem um dos quatro rótulos de classe:

Uma árvore de decisão simples construída sobre esses dados dividirá iterativamente os dados ao longo de um ou outro eixo de acordo com algum critério quantitativo e, em cada nível, atribuirá o rótulo da nova região de acordo com a maioria dos votos de pontos dentro dela.

Esse processo de ajustar uma árvore de decisão aos nossos dados pode ser feito no Scikit-Learn com o `DecisionTreeClassifier` estimador:

````python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier().fit(X, y)
````

Vamos escrever uma função de utilidade rápida para nos ajudar a visualizar a saída do classificador:

````python
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()

    # Plot the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=cmap,
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # fit the estimator
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)


    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap=cmap, clim=(y.min(), y.max()),
                           zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
````

````python
visualize_classifier(DecisionTreeClassifier(), X, y)
````

Observe que à medida que a profundidade aumenta, tendemos a obter regiões de classificação com formas muito estranhas

##Árvores de decisão e over-fitting

Esse superajuste acaba sendo uma propriedade geral das árvores de decisão: é muito fácil ir muito fundo na árvore e, assim, ajustar os detalhes dos dados específicos em vez das propriedades gerais das distribuições das quais eles são extraídos. Outra maneira de ver esse ajuste excessivo é observar os modelos treinados em diferentes subconjuntos de dados - por exemplo, treinar duas árvores diferentes, cada uma na metade dos dados originais.

#Conjuntos de Estimadores: Florestas Aleatórias

Essa noção – que vários estimadores de superajuste podem ser combinados para reduzir o efeito desse superajuste – é a base de um método de conjunto chamado bagging . O ensacamento faz uso de um conjunto (uma sacola, talvez) de estimadores paralelos, cada um dos quais superajusta os dados e calcula a média dos resultados para encontrar uma classificação melhor. Um conjunto de árvores de decisão aleatórias é conhecido como floresta aleatória.

Esse tipo de classificação de ensacamento pode ser feito manualmente usando `BaggingClassifier` o meta-estimador do Scikit-Learn

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators=100, max_samples=0.8,
                        random_state=1)

bag.fit(X, y)
visualize_classifier(bag, X, y)
````

 Florestas aleatórias também podem funcionar no caso de regressão (isto é, variáveis ​​contínuas em vez de categóricas). O estimador a ser usado para isso é o RandomForestRegressor, e a sintaxe é muito parecida com a que vimos anteriormente.

````python
rng = np.random.RandomState(42)
x = 10 * rng.rand(200)

def model(x, sigma=0.3):
    fast_oscillation = np.sin(5 * x)
    slow_oscillation = np.sin(0.5 * x)
    noise = sigma * rng.randn(len(x))

    return slow_oscillation + fast_oscillation + noise

y = model(x)
plt.errorbar(x, y, 0.3, fmt='o');
````

Usando o regressor da floresta aleatória, podemos encontrar a melhor curva de ajuste da seguinte forma:

````python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(200)
forest.fit(x[:, None], y)

xfit = np.linspace(0, 10, 1000)
yfit = forest.predict(xfit[:, None])
ytrue = model(xfit, sigma=0)

plt.errorbar(x, y, 0.3, fmt='o', alpha=0.5)
plt.plot(xfit, yfit, '-r');
plt.plot(xfit, ytrue, '-k', alpha=0.5);
````

Aqui, o modelo verdadeiro é mostrado na curva cinza suave, enquanto o modelo de floresta aleatória é mostrado na curva vermelha irregular. Como você pode ver, o modelo de floresta aleatória não paramétrica é flexível o suficiente para ajustar os dados de vários períodos, sem a necessidade de especificar um modelo de vários períodos!

#Em Profundidade: Análise de Componentes Principais

O PCA é fundamentalmente um algoritmo de redução de dimensionalidade, mas também pode ser útil como uma ferramenta para visualização, para filtragem de ruído, para extração de recursos e engenharia e muito mais. Após uma breve discussão conceitual do algoritmo PCA, veremos alguns exemplos dessas outras aplicações.

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
````

A análise de componentes principais é um método não supervisionado rápido e flexível para redução de dimensionalidade em dados. Seu comportamento é mais fácil de visualizar observando um conjunto de dados bidimensional. Considere os seguintes 200 pontos:

````python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
````

a configuração do problema aqui é um pouco diferente do problema de regressão linear, em vez de tentar prever os valores de y a partir dos valores de x, o problema de aprendizado não supervisionado tenta aprender sobre a relação entre os valores x e y.
Na análise de componentes principais, esse relacionamento é quantificado encontrando uma lista dos eixos principais nos dados e usando esses eixos para descrever o conjunto de dados. Usando o estimador do Scikit-Learn PCA, podemos calcular isso da seguinte forma:

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
````

O ajuste aprende algumas quantidades dos dados, principalmente os "componentes" e a "variância explicada":

PCA como redução de dimensionalidade
O uso de PCA para redução de dimensionalidade envolve zerar um ou mais dos menores componentes principais, resultando em uma projeção de dimensão inferior dos dados que preserva a variância máxima dos dados.

Aqui está um exemplo de uso do PCA como uma transformada de redução de dimensionalidade:

````python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
````

Os pontos claros são os dados originais, enquanto os pontos escuros são a versão projetada. Isso deixa claro o que significa uma redução de dimensionalidade PCA: a informação ao longo do eixo ou eixos principais menos importantes é removida, deixando apenas o(s) componente(s) dos dados com a maior variância. A fração de variação que é cortada (proporcional à dispersão de pontos sobre a linha formada nesta figura) é aproximadamente uma medida de quanta "informação" é descartada nesta redução de dimensionalidade.

##RESUMO

A aprendizagem múltipla, também conhecida como ensemble learning, é uma abordagem no campo do aprendizado de máquina em que vários modelos individuais são combinados para formar um modelo mais poderoso e robusto. Em vez de confiar em um único modelo, a aprendizagem múltipla busca aproveitar a diversidade e a complementaridade dos modelos componentes para melhorar o desempenho preditivo.

Existem diferentes técnicas de aprendizagem múltipla, mas a ideia geral é construir um conjunto (ensemble) de modelos que sejam treinados independentemente e, em seguida, combinar suas previsões para chegar a uma previsão final. Alguns dos métodos de aprendizagem múltipla mais comuns incluem:

* **Método de média:** Nesse método, os modelos componentes geram previsões individuais e, em seguida, a média dessas previsões é tomada como a previsão final. Isso é aplicado tanto para problemas de classificação (onde é chamado de voto majoritário) quanto para problemas de regressão (onde é chamado de média dos valores previstos).

* **Método de votação:** Esse método é semelhante ao método de média, mas em vez de tomar a média das previsões, é feita uma votação para determinar a previsão final. Cada modelo componentes vota em uma classe ou valor previsto, e a classe ou valor mais votado é escolhido como a previsão final.

* **Método de ponderação:** Nesse método, cada modelo componente recebe um peso que reflete sua confiabilidade ou desempenho. As previsões de cada modelo são ponderadas pelos respectivos pesos antes de serem combinadas para formar a previsão final. A atribuição de pesos pode ser baseada em métricas de desempenho no conjunto de treinamento ou validação.

* **Método de empilhamento (stacking):** Esse método envolve o treinamento de um modelo adicional, chamado de modelo meta-aprendiz (meta-learner), que aprende a combinar as previsões dos modelos componentes. Os modelos componentes geram suas previsões, que são usadas como entrada para o modelo meta-aprendiz, que então gera a previsão final.

A aprendizagem múltipla tem várias vantagens. Ela pode melhorar a precisão e a generalização do modelo, reduzir o overfitting e lidar com a variabilidade nos dados. Além disso, a combinação de modelos pode compensar as fraquezas individuais de cada modelo componente.

No entanto, a aprendizagem múltipla também requer cuidados na seleção dos modelos componentes, na diversidade dos modelos e na forma como as previsões são combinadas. É importante que os modelos componentes sejam diferentes o suficiente para fornecer perspectivas complementares, mas não tão diferentes a ponto de se tornarem irreconciliáveis.

Em resumo, a aprendizagem múltipla é uma abordagem poderosa no aprendizado de máquina que combina vários modelos individuais para melhorar a precisão e a robustez das previsões. Ela é amplamente utilizada em problemas de classificação, regressão e outras tarefas de aprendizado de máquina.

##Algoritmo de k-médias: Expectativa–Maximização ¶
Expectation-maximization (E-M) é um algoritmo poderoso que surge em uma variedade de contextos dentro da ciência de dados. k -means é uma aplicação do algoritmo particularmente simples e fácil de entender, e vamos abordá-la brevemente aqui. Em suma, a abordagem de maximização de expectativa aqui consiste no seguinte procedimento:

Adivinhe alguns centros de cluster
Repita até convergir
* E-Step : atribua pontos ao centro de cluster mais próximo
* M-Step : definir os centros de cluster para a média


Aqui, o "passo E" ou "passo de expectativa" é assim chamado porque envolve a atualização de nossa expectativa de qual cluster cada ponto pertence. O "passo M" ou "passo de maximização" é assim chamado porque envolve a maximização de alguma função de aptidão que define a localização dos centros do cluster - neste caso, essa maximização é realizada tomando uma média simples dos dados em cada cluster .

A literatura sobre esse algoritmo é vasta, mas pode ser resumida da seguinte forma: em circunstâncias típicas, cada repetição do passo-E e do passo-M sempre resultará em uma melhor estimativa das características do cluster.

## Resumo do K-Means

Podemos resumir esse algoritmo da seguinte forma:

O algoritmo k-means é um método de aprendizado não supervisionado utilizado para agrupar um conjunto de dados em k grupos distintos. O objetivo é encontrar k centroides que representem os centros dos grupos, de forma que a soma das distâncias entre os pontos e seus centroides correspondentes seja minimizada.

Aqui está uma descrição passo a passo do algoritmo k-means:

* Inicialização: Escolha k centroides iniciais aleatoriamente ou de forma determinística a partir dos dados.

* Atribuição: Atribua cada ponto de dados ao centroide mais próximo com base na distância euclidiana (ou outra medida de distância). Isso cria k grupos iniciais.

* Atualização: Recalcule os centroides de cada grupo, calculando a média das coordenadas dos pontos atribuídos a cada grupo.

* Reatribuição: Repita os passos 2 e 3 até que os centroides não mudem significativamente ou até que um critério de parada seja atingido (por exemplo, um número máximo de iterações).

* Convergência: Os centroides convergem para as posições finais e cada ponto de dados é atribuído ao seu grupo correspondente.


O número de grupos, k, é um parâmetro definido pelo usuário e geralmente é determinado com base no conhecimento prévio do problema ou por meio de técnicas de validação cruzada.

O algoritmo k-means tem algumas limitações e desafios. Ele pode convergir para um mínimo local, o que significa que a solução encontrada pode não ser a ótima. Além disso, é sensível à inicialização dos centroides, o que pode levar a diferentes soluções finais. Existem variações do algoritmo, como o k-means++, que abordam essas questões.

O k-means é amplamente utilizado em várias áreas, como segmentação de clientes, análise de agrupamento de dados e processamento de imagens. É um algoritmo eficiente e escalável, embora sua eficácia possa ser afetada por dados de alta dimensionalidade e pela presença de outliers.

É importante ressaltar que o algoritmo k-means é um método heurístico e não garante a melhor solução global. Portanto, é recomendado realizar análise adicional e interpretação dos resultados para avaliar a qualidade dos agrupamentos obtidos.

##Generalizando E–M: Modelos de Mistura Gaussiana

Um modelo de mistura gaussiana (GMM) tenta encontrar uma mistura de distribuições de probabilidade gaussianas multidimensionais que melhor modelam qualquer conjunto de dados de entrada. No caso mais simples, os GMMs podem ser usados ​​para encontrar clusters da mesma maneira que k -means:

````python
from sklearn.mixture import GaussianMixture as GMM
gmm = GMM(n_components=4).fit(X)
labels = gmm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis');
````

Mas como o GMM contém um modelo probabilístico sob o capô, também é possível encontrar atribuições de cluster probabilístico - no Scikit-Learn, isso é feito usando o `predict_proba` método. Isso retorna uma matriz de tamanho `[n_samples, n_clusters]` que mede a probabilidade de que qualquer ponto pertença ao cluster fornecido, extremamente util para avaliar a sobreposição de dados:

````python
probs = gmm.predict_proba(X)
print(probs[:5].round(3))
````

Podemos visualizar essa incerteza, por exemplo, tornando o tamanho de cada ponto proporcional à certeza de sua previsão; olhando para a figura a seguir, podemos ver que são precisamente os pontos nos limites entre os clusters que refletem essa incerteza de atribuição de cluster:

````python
size = 50 * probs.max(1) ** 2  # square emphasizes differences
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=size);
````

Sob o capô, um modelo de mistura gaussiana é muito semelhante ao k -means: ele usa uma abordagem de maximização de expectativa que qualitativamente faz o seguinte:

* Escolha palpites iniciais para a localização e forma

* Repita até convergir:

* * E-etapa : para cada ponto, encontre os pesos que codificam a probabilidade de associação em cada cluster;
* * Etapa M : para cada cluster, atualize sua localização, normalização e forma com base em todos os pontos de dados, fazendo uso dos pesos.

O resultado disso é que cada aglomerado está associado não a uma esfera de arestas rígidas, mas a um modelo gaussiano suave. Assim como na abordagem de maximização de expectativa k -means, esse algoritmo pode às vezes perder a solução globalmente ótima e, portanto, na prática, várias inicializações aleatórias são usadas.

Vamos criar uma função que nos ajudará a visualizar as localizações e formas dos clusters do GMM desenhando elipses com base na saída do GMM:

````python
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()

    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)

    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))

def plot_gmm(gmm, X, label=True, ax=None):
    ax = ax or plt.gca()
    labels = gmm.fit(X).predict(X)
    if label:
        ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
    else:
        ax.scatter(X[:, 0], X[:, 1], s=40, zorder=2)
    ax.axis('equal')

    w_factor = 0.2 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        draw_ellipse(pos, covar, alpha=w * w_factor)
````

````python
gmm = GMM(n_components=4, random_state=42)
plot_gmm(gmm, X)
````

##Resumo da seção

O algoritmo de mistura gaussiana (Gaussian Mixture Algorithm) é uma técnica de aprendizado de máquina que envolve a modelagem de dados como uma combinação de várias distribuições gaussianas. É um método de aprendizado não supervisionado usado para realizar a clusterização dos dados, ou seja, agrupá-los em diferentes grupos ou clusters com base em suas características.

A ideia básica por trás do algoritmo de mistura gaussiana é que cada cluster é representado por uma distribuição gaussiana, e os dados são gerados por uma combinação dessas distribuições. O objetivo do algoritmo é estimar os parâmetros dessas distribuições gaussianas, como as médias, as matrizes de covariância e os pesos relativos de cada uma.

O algoritmo de mistura gaussiana segue um processo iterativo. Inicialmente, são gerados aleatoriamente os parâmetros iniciais para cada distribuição gaussiana. Em seguida, os dados são atribuídos a cada cluster com base nas probabilidades calculadas usando as distribuições gaussianas. Posteriormente, os parâmetros das distribuições são atualizados com base nas atribuições dos dados. Esse processo é repetido até que ocorra a convergência, ou seja, até que os parâmetros estimados não mudem significativamente.

Uma vez que o algoritmo de mistura gaussiana tenha convergido, cada ponto de dados é atribuído ao cluster correspondente com base na distribuição com maior probabilidade para esse ponto. Isso permite a realização da clusterização dos dados.

O algoritmo de mistura gaussiana é amplamente utilizado em várias aplicações, como reconhecimento de padrões, análise de dados, visão computacional e processamento de linguagem natural. Ele oferece uma abordagem flexível para modelar dados complexos que não seguem uma distribuição simples, permitindo a descoberta de estruturas e padrões ocultos nos dados.

#A estimativa de densidade do kernel (KDE)

A estimativa de densidade do kernel (Kernel Density Estimation - KDE) é uma técnica utilizada para estimar a função de densidade de probabilidade subjacente a um conjunto de dados. Ela é frequentemente usada em análise exploratória de dados e em várias aplicações de aprendizado de máquina.

A ideia principal do KDE é suavizar os dados para obter uma estimativa contínua da densidade de probabilidade. Em vez de considerar cada ponto de dados individualmente, o KDE considera uma janela (kernel) em torno de cada ponto e calcula a contribuição desse kernel na densidade em cada ponto ao longo do domínio dos dados.

O kernel é uma função de densidade de probabilidade simétrica e não negativa, centrada em torno de cada ponto de dados. Ele determina a forma da contribuição daquele ponto para a estimativa de densidade em sua vizinhança. Exemplos comuns de kernels utilizados no KDE incluem o kernel gaussiano (também conhecido como normal) e o kernel uniforme.

A estimação da densidade de probabilidade em um determinado ponto é obtida somando as contribuições dos kernels centrados em todos os pontos de dados, ponderadas pela distância entre o ponto de interesse e os pontos de dados. Quanto mais próximos os pontos de dados estiverem do ponto de interesse, maior será sua contribuição para a estimativa de densidade nesse ponto.

O parâmetro crucial no KDE é a largura de banda (bandwidth), que controla o tamanho do kernel e a influência dos pontos de dados vizinhos. Uma largura de banda pequena produz uma estimativa mais detalhada e "ruidosa" da densidade, enquanto uma largura de banda grande suaviza a estimativa, reduzindo o detalhe.

O KDE é uma técnica versátil e pode ser usada em várias áreas, como visualização de dados, detecção de anomalias, geração de amostras e modelagem estatística. Ele fornece uma representação contínua da densidade de probabilidade, permitindo uma compreensão mais intuitiva da distribuição subjacente aos dados.

Estimativa de densidade refere-se a uma técnica usada para estimar a função de densidade de probabilidade (PDF) subjacente a um conjunto de dados. A densidade de probabilidade representa a distribuição de probabilidade de uma variável aleatória contínua.

A estimativa de densidade é útil para entender a distribuição dos dados e identificar padrões e características importantes. Ela é amplamente utilizada em análise exploratória de dados, visualização, modelagem estatística e várias aplicações de aprendizado de máquina.

A estimativa de densidade geralmente envolve a escolha de um modelo paramétrico ou não paramétrico. Os modelos paramétricos assumem uma forma específica para a função de densidade de probabilidade, como a distribuição gaussiana (normal) ou a distribuição exponencial. Nesses casos, os parâmetros do modelo são estimados a partir dos dados para ajustar a distribuição aos dados observados.

Por outro lado, os modelos não paramétricos, como a estimativa de densidade do kernel (KDE), não fazem suposições sobre a forma funcional da distribuição subjacente. Eles fornecem uma estimativa suave da densidade de probabilidade, geralmente usando uma função de kernel para suavizar os dados e obter uma representação contínua da densidade.

A estimativa de densidade pode ser usada para várias finalidades, como identificar a presença de múltiplos picos (modos) na distribuição, detectar anomalias, gerar amostras de dados sintéticos, realizar testes de hipóteses estatísticas e muito mais. Ela ajuda a capturar a estrutura subjacente dos dados e pode fornecer insights valiosos sobre o comportamento dos dados.

Abaixo, temos um exemplo com um kernel gaussiano e uma largura de banda de 0.5. Em seguida, ajustamos o modelo aos dados fornecidos. Em seguida, geramos pontos ao longo de um intervalo específico e usamos o modelo KDE para estimar a densidade em cada um desses pontos. Por fim, plotamos a estimativa de densidade usando a biblioteca Matplotlib.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Dados de exemplo
data = np.array([2.5, 3.1, 2.7, 1.9, 3.5, 4.2, 3.8, 4.9, 3.6, 2.8]).reshape(-1, 1)

# Criação do modelo KDE
kde = KernelDensity(kernel='gaussian', bandwidth=0.5)

# Ajuste do modelo aos dados
kde.fit(data)

# Geração de pontos para estimativa da densidade
x = np.linspace(0, 6, 500).reshape(-1, 1)

# Estimativa de densidade usando o modelo KDE
log_dens = kde.score_samples(x)

# Plot da estimativa de densidade
plt.fill_between(x[:, 0], np.exp(log_dens), alpha=0.5)
plt.plot(data[:, 0], np.full_like(data[:, 0], -0.01), '|k', markeredgewidth=1)
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.title('Estimativa de Densidade do Kernel (KDE)')
plt.show()
````
